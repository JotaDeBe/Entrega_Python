1. Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias
de cada letra en la cadena. Los espacios no deben ser considerados.

```python
def frecuencias_letras(cadena):
    """
    Summary:
        Función que calcula la frecuencia de cada letra dentro de una cadena de texto,
        ignorando los espacios.

    Args:
        cadena (str): Texto del cual se contarán las letras.

    Returns:
        dict: Diccionario donde las claves son las letras y los valores
              representan cuántas veces aparece cada una en la cadena.
    """

    cadena = cadena.replace(" ", "")  
    diccionario = {}

    for letra in cadena:
        if letra in diccionario:
            diccionario[letra] += 1  
        else:
            diccionario[letra] = 1 
    return diccionario
```

2. Dada una lista de números, obtén una nueva lista con el doble de cada valor. Usa la función map()

3. Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe
devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo. 


```python
def filtrar_palabras(lista, objetivo):
    """
    Summary:
        Filtra y devuelve las palabras que contienen la palabra objetivo.

    Args:
        lista (list): Lista de palabras (strings).
        objetivo (str): Palabra que se desea buscar dentro de las palabras de la lista.

    Returns:
        list: Nueva lista con palabras que contienen la palabra objetivo.
    """
    nueva_lista = []

    for palabra in lista:
        if objetivo in palabra:  
            nueva_lista.append(palabra)

    return nueva_lista

```

5. Ecribe una función que tome una lista de números como parámetro y un valor opcional nota_aprobado, que por
defecto es 5. La función debe calcular la media de los números en la lista y determinar si la media es mayor o igual
que nota aprobado. Si es así, el estado será "aprobado", de lo contrario, será "suspenso". La función debe devolver
una tupla que contenga la media y el estado.


```python
def media(lista, nota_aprobado=5):
    """
    Summary:
        Calcula la media de una lista de números y determina si está aprobada.

    Args:
        lista (list): Lista de números.
        nota_aprobado (int, optional): Nota mínima para aprobar. Por defecto 5.

    Returns:
        tuple: (media, "aprobado" o "suspenso")
    """
    promedio = sum(lista) / len(lista)  

    if promedio >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"

    return (promedio, estado)

   
```

6. Escribe una función que calcule el factorial de un número de manera recursiva.


```python
def factorial(n):
    """
    Calcula el factorial de un número de manera recursiva.
    """
    if n == 0 or n == 1:  
        return 1
    else:
        return n * factorial(n - 1) 
```

10. Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una
excepción personalizada y maneja el error adecuadamente.



```python
class ListaVaciaError(Exception):
    """
    Excepción personalizada para indicar que la lista de números
    está vacía y no se puede calcular el promedio.
    """
    pass


def promedio(lista):
    """
    Summary:
        Calcula el promedio de una lista de números. Si la lista está vacía,
        lanza una excepción personalizada ListaVaciaError y la maneja mostrando
        un mensaje.

    Args:
        lista (list): Lista de números (int o float) de la que se calculará el promedio.

    Returns:
        float | None: Devuelve el promedio de los números como float si la lista
                      no está vacía. Devuelve None si la lista está vacía.
    """
    try:
        if len(lista) == 0:
            raise ListaVaciaError("La lista está vacía, no se puede calcular el promedio.")
        
        return sum(lista) / len(lista)
    
    except ListaVaciaError:
        print(ListaVaciaError)
        return None


```

mayúsculas y minúsculas. Las letras no pueden estar repetidas .Usa la función map()


```python
def letras_mayus_minus(caracteres):
    """
    Summary:
        Genera una lista de tuplas donde cada tupla contiene una letra en mayúscula
        y minúscula. Los caracteres repetidos son eliminados.

    Args:
        caracteres (iterable): Conjunto de caracteres, puede ser un string o lista.

    Returns:
        list: Lista de tuplas con el formato (MAYÚSCULA, minúscula) sin repetidos.
    """
    unicas = set(caracteres)
    return list(map(lambda c: (c.upper(), c.lower()), unicas))
```

28. Crea una función que busque y devuelva el primer elemento duplicado en una lista dada.

```python
def duplicado(lista):
    """
    Summary:
        Busca y devuelve el primer elemento duplicado en una lista.

    Args:
        lista (list): Lista de elementos donde se buscarán duplicados.

    Returns:
        El primer elemento duplicado encontrado en la lista, o None si no hay duplicados.
    """
    vistos = set()
    for e in lista:
        if e in vistos:
            return e
        vistos.add(e)
    return None

```

29. Crea una función que convierta una variable en una cadena de texto y enmascare todos los caracteres con el
carácter '#', excepto los últimos cuatro.

```python
def enmascarar(variable):
    """
    Enmascara todos los caracteres de una variable convertida a cadena
    con el carácter '#' excepto los últimos cuatro.
    
    Parámetros:
        variable: cualquier tipo (int, str, float, etc.)
    
    Retorna:
        Una cadena en la que todos los caracteres excepto los últimos cuatro
        están reemplazados por '#'.
    """
    s = str(variable)  

    if len(s) <= 4:
        return s  

    ocultar = "#" * (len(s) - 4)  
    ultimos = s[-4:]  
    
    return ocultar + ultimos

```

30. Crea una función que determine si dos palabras son anagramas, es decir, si están formadas por las mismas letras
pero en diferente orden.


```python
def son_anagramas(palabra1, palabra2):
    """
    Determina si dos palabras son anagramas.
    
    Args:
        palabra1 (str): Primera palabra.
        palabra2 (str): Segunda palabra.
        
    Returns:
        bool: True si son anagramas, False en caso contrario.
    """
    
    p1 = palabra1.lower()
    p2 = palabra2.lower()
    
    return sorted(p1) == sorted(p2)
```

34. Crea la clase Arbol , define un árbol genérico con un tronco y ramas como atributos. Los métodos disponibles son:
crecer_tronco , nueva_rama , crecer_ramas , quitar_rama e info_arbol . El objetivo es implementar estos métodos para
manipular la estructura del árbol.
Código a seguir:
1. Inicializar un árbol con un tronco de longitud 1 y una lista vacía de ramas.
2. Implementar el método crecer_tronco para aumentar la longitud del tronco en una unidad.
3. Implementar el método nueva_rama para agregar una nueva rama de longitud 1 a la lista de ramas.
4. Implementar el método crecer_ramas para aumentar en una unidad la longitud de todas las ramas existentes.
5. Implementar el método quitar_rama para eliminar una rama en una posición específica.
6. Implementar el método
info_arbol para devolver información sobre la longitud del tronco, el número de ramas y las longitudes de las
mismas.
Caso de uso:
1. Crear un árbol.
2. Hacer crecer el tronco del árbol una unidad.
3. Añadir una nueva rama al árbol.
4. Hacer crecer todas las ramas del árbol una unidad.
5. Añadir dos nuevas ramas al árbol.
6. Retirar la rama situada en la posición 2.
7. Obtener información sobre el árbol.


```python
class Arbol:
    def __init__(self):
        self.tronco = 1
        self.ramas = []

    def crecer_tronco(self):
        """
        Aumenta la longitud del tronco en una unidad.
        """
        self.tronco += 1

    def nueva_rama(self):
        """
        Añade una nueva rama con longitud 1.
        """
        self.ramas.append(1)

    def crecer_ramas(self):
        """
        Aumenta en una unidad la longitud de todas las ramas existentes.
        """
        for i in range(len(self.ramas)):
            self.ramas[i] += 1

    def quitar_rama(self, posicion):
        """
        Elimina la rama situada en la posición indicada.
        Se comprueba que la posición exista.
        """
        if 0 <= posicion < len(self.ramas):
            del self.ramas[posicion]
        else:
            print("Posición no válida, no se puede eliminar la rama.")

    def info_arbol(self):
        """
        Devuelve información completa del árbol:
        longitud del tronco, número de ramas y longitudes de cada una.
        """
        return {
            "Longitud del tronco": self.tronco,
            "Número de ramas": len(self.ramas),
            "Longitudes de ramas": self.ramas
        }


# ---------- CASO DE USO ----------

# 1. Crear el árbol
arbol = Arbol()

# 2. Hacer crecer el tronco
arbol.crecer_tronco()

# 3. Añadir una rama
arbol.nueva_rama()

# 4. Hacer crecer todas las ramas
arbol.crecer_ramas()

# 5. Añadir dos ramas nuevas
arbol.nueva_rama()
arbol.nueva_rama()

# 6. Quitar la rama en posición 2
arbol.quitar_rama(2)

# 7. Obtener información final del árbol
info = arbol.info_arbol()
print(info)
```

```python
class UsuarioBanco:
    def __init__(self, nombre, saldo, cuenta_corriente):
        """
        Inicializa un usuario del banco.
        nombre: str
        saldo: float o int
        cuenta_corriente: bool (True/False)
        """
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):
        """
        Retira dinero del saldo del usuario.
        Lanza error si no hay suficiente saldo.
        """
        if cantidad > self.saldo:
            raise ValueError(f"{self.nombre} no tiene saldo suficiente para retirar {cantidad}")
        self.saldo -= cantidad

    def agregar_dinero(self, cantidad):
        """
        Agrega dinero al saldo del usuario.
        """
        self.saldo += cantidad

    def transferir_dinero(self, desde_usuario, cantidad):
        """
        Transfiere dinero desde otro usuario al usuario actual.
        Lanza error si el otro usuario no tiene suficiente saldo.
        """
        if cantidad > desde_usuario.saldo:
            raise ValueError(f"{desde_usuario.nombre} no tiene saldo suficiente para transferir {cantidad}")
        desde_usuario.saldo -= cantidad
        self.saldo += cantidad

# ---------- CASO DE USO ----------

# 1. Crear dos usuarios
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)

# 2. Agregar 20 unidades al saldo de Bob
bob.agregar_dinero(20)  # Bob saldo = 70

# 3. Hacer una transferencia de 80 unidades desde Bob a Alicia
try:
    alicia.transferir_dinero(bob, 80)  # Alice recibe 80, Bob pierde 80
except ValueError as e:
    print(e)

# 4. Retirar 50 unidades de saldo de Alicia
try:
    alicia.retirar_dinero(50)
except ValueError as e:
    print(e)

# Información final
print(f"Alicia saldo final: {alicia.saldo}")
print(f"Bob saldo final: {bob.saldo}")
```

37. Crea una función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras ,
reemplazar_palabras , eliminar_palabra . Estas opciones son otras funciones que tenemos que definir primero y llamar dentro
de la función procesar_texto .
Código a seguir:
1. Crear una función contar_palabras para contar el número de veces que aparece cada palabra en el texto. Tiene
que devolver un diccionario.
2. Crear una función reemplazar_palabras para remplazar una palabra_original del texto por una palabra_nueva . Tiene
que devolver el texto con el remplazo de palabras.
3. Crear una función eliminar_palabra para eliminar una palabra del texto. Tiene que devolver el texto con la palabra
eliminada.
4. Crear la función procesar_texto que tome un texto, una opción(entre "contar", "reemplazar", "eliminar") y un
número de argumentos variable según la opción indicada.
Caso de uso:
Comprueba el funcionamiento completo de la función procesar_texto


```python
# ---------- FUNCIONES AUXILIARES ----------

def contar_palabras(texto):
    """
    Cuenta cuántas veces aparece cada palabra en el texto.
    Devuelve un diccionario {palabra: cantidad}.
    """
    palabras = texto.split()  
    contador = {}
    for palabra in palabras:
        contador[palabra] = contador.get(palabra, 0) + 1
    return contador

def reemplazar_palabras(texto, palabra_original, palabra_nueva):
    """
    Reemplaza todas las ocurrencias de palabra_original por palabra_nueva.
    Devuelve el texto modificado.
    """
    return texto.replace(palabra_original, palabra_nueva)

def eliminar_palabra(texto, palabra):
    """
    Elimina todas las ocurrencias de una palabra del texto.
    Devuelve el texto modificado.
    """
    texto_modificado = texto.replace(palabra, "")
    texto_modificado = ' '.join(texto_modificado.split())
    return texto_modificado

# ---------- FUNCIÓN PRINCIPAL ----------

def procesar_texto(texto, opcion, *args):
    """
    Procesa el texto según la opción indicada.
    opcion: "contar", "reemplazar", "eliminar"
    *args: argumentos adicionales según la opción
    """
    if opcion == "contar":
        return contar_palabras(texto)
    elif opcion == "reemplazar":
        if len(args) != 2:
            raise ValueError("Para 'reemplazar' se requieren dos argumentos: palabra_original, palabra_nueva")
        return reemplazar_palabras(texto, args[0], args[1])
    elif opcion == "eliminar":
        if len(args) != 1:
            raise ValueError("Para 'eliminar' se requiere un argumento: palabra")
        return eliminar_palabra(texto, args[0])
    else:
        raise ValueError("Opción no válida. Debe ser 'contar', 'reemplazar' o 'eliminar'.")

# ---------- EJEMPLO DE USO ----------

texto = "hola mundo hola python mundo"
print(procesar_texto(texto, "contar"))  # {'hola': 2, 'mundo': 2, 'python': 1}
print(procesar_texto(texto, "reemplazar", "mundo", "universo"))  # 'hola universo hola python universo'
print(procesar_texto(texto, "eliminar", "hola"))  # 'mundo python mundo'
```

38. Genera un programa que nos diga si es de noche, de día o tarde según la hora proporcionada por el usuario.


```python
def horario(hora_str):
    """
    Determina si es de noche, de día o tarde según la hora.
    hora_str: cadena en formato "HH:MM"
    """
    try:
        partes = hora_str.split(":")
        if len(partes) != 2:
            raise ValueError("Formato incorrecto, usar HH:MM")
        
        hora = int(partes[0])
        minutos = int(partes[1])
        
        
        if not (0 <= hora <= 23) or not (0 <= minutos <= 59):
            raise ValueError("Hora o minuto fuera de rango")
        
        
        hora_decimal = hora + minutos / 60
        
        
        if 6 <= hora_decimal < 12:
            print("Es de día")
        elif 12 <= hora_decimal < 20:
            print("Es tarde")
        else:
            print("Es de noche")
            
    except ValueError as e:
        print("Error:", e)
```

40. Escribe una función que tome dos parámetros: figura (una cadena que puede ser "rectangulo" , "circulo" o
"triangulo" ) y datos (una tupla con los datos necesarios para calcular el área de la figura).


```python
import math

def area(figura, datos):
    """
    Calcula el área de una figura geométrica.
    
    figura: str → "rectangulo", "triangulo" o "circulo"
    datos: tupla → valores necesarios para calcular el área
           Rectángulo y Triángulo: (base, altura)
           Círculo: (radio,)
    
    Devuelve: área como float
    """
    if figura == "rectangulo":
        base, altura = datos
        return base * altura
    elif figura == "triangulo":
        base, altura = datos
        return (base * altura) / 2
    elif figura == "circulo":
        (radio,) = datos  
        return math.pi * radio**2
    else:
        raise ValueError("Figura no válida. Debe ser 'rectangulo', 'triangulo' o 'circulo'.")
```

41. En este ejercicio, se te pedirá que escribas un programa en Python que utilice condicionales para determinar el
monto final de una compra en una tienda en línea, después de aplicar un descuento. El programa debe hacer lo
siguiente:
1. Solicita al usuario que ingrese el precio original de un artículo.
2. Pregunta al usuario si tiene un cupón de descuento (respuesta sí o no).
3. Si el usuario responde que sí, solicita que ingrese el valor del cupón de descuento.
4. Aplica el descuento al precio original del artículo, siempre y cuando el valor del cupón sea válido (es decir, mayor
a cero). Por ejemplo, descuento de 15€.
5. Muestra el precio final de la compra, teniendo en cuenta el descuento aplicado o sin él.
6. Recuerda utilizar estructuras de control de flujo como if, elif y else para llevar a cabo estas acciones en tu
programa de Python.

```python
def calcular_precio_final():
    """
    Calcula el precio final de una compra en una tienda en línea
    considerando un posible cupón de descuento.
    """
    try:
        precio_original = float(input("Ingresa el precio original del artículo: "))
        if precio_original < 0:
            print("El precio no puede ser negativo.")
            return
        
        respuesta = input("¿Tienes un cupón de descuento? (sí/no): ").strip().lower()
        
        precio_final = precio_original
        
        if respuesta == "sí":
            descuento = float(input("Ingresa el valor del cupón de descuento: "))
            
            if descuento > 0:
                precio_final -= descuento
            else:
                print("El valor del cupón no es válido. No se aplicará descuento.")
        
        print(f"El precio final de la compra es: {precio_final:.2f} €")
    
    except ValueError:
        print("Error: Debes ingresar un número válido para el precio o el cupón.")

calcular_precio_final()
```
